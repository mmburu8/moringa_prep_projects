```
import pandas as pd
import numpy as np
```

 **LOADING CELLS_GEO.CSV AND SPLITTING COLUMNS** 

```
df = pd.read_csv("cells_geo.csv")
df.columns = ["total"]
#splitting the columns
df = df.total.str.split(";",expand=True)
df
```

**RENAMING OF COLUMNS**

```
df.columns = ["NUMBERS", "VILLES", "STATUS", "LOCALISATION", "DECOUPZONE", "ZONENAME", "LONGITUDE", "LATITUDE", "REGION", "AREA", "CELL_ID", "SITE_CODE",]
df.head(30)
```

**DELETE ROWS WHERE CELL_ID IS NOT IN SERVICE**

```
df.drop(df[df["STATUS"] != "In Service"].index, inplace=True)
df.head(30)
```

**DROPPING OF NULL AND DUPLICATE VALUES IN DF TABLE**

```
df.duplicated()
df_d = df.drop_duplicates()
df_d
```

```
df_d.isnull()
df_d.dropna(how= 'all')
df_d.fillna(0)
df = df_d
df
```

**LOADING TELCOM DATASETS, RENAMING NAME OF COLUMNS AND MERGING DATASETS**

```
day_1 = pd.read_csv("Telcom_dataset.csv")
day_1.rename(columns = {"PRODUTC": "PRODUCT"}, inplace = True)
day_1.rename(columns = {"DATETIME": "DATE_TIME"}, inplace = True)
day_2 = pd.read_csv("Telcom_dataset2.csv")
day_2.rename(columns = {"DW_A_NUMBER": "DW_A_NUMBER_INT"}, inplace = True)
day_2.rename(columns = {"DW_B_NUMBER": "DW_B_NUMBER_INT"}, inplace = True)
day_3 = pd.read_csv("Telcom_dataset3.csv")
day_3.rename(columns = {"CELLID": "CELL_ID"}, inplace = True)
day_3.rename(columns = {"SIET_ID": "SITE_ID"}, inplace = True)
all_days = pd.concat((day_1, day_2, day_3), axis= 0, ignore_index = True)
#group_values = all_days["VALUE"].groupby(all_days["CELL_ID"])
all_days
```

**DELETE IRRELEVANT AND REDUNDANT COLUMNS**

```
all_days = all_days.drop(columns = ["DW_A_NUMBER_INT", "DW_B_NUMBER_INT", "CELL_ON_SITE"])
all_days.head(10)
```

**DROPPING OF NULL AND DUPLICATE VALUES IN ALL_DAYS TABLE**

```
all_days.duplicated()
all_days_d = all_days.drop_duplicates()
all_days_d
```

```
all_days_d.isnull()
all_days_d.dropna(how= 'all')
all_days_d.fillna(0)
all_days = all_days_d
all_days
```

**RANK OF PRODUCTS ACCORDING TO SUM OF THEIR VALUES**

```
product_group = all_days.groupby(["PRODUCT"])["VALUE"].sum()
product_group.sort_values(ascending=False)
```

**NUMBER OF CELL_IDS IN EACH CITY**

```
group_CELLID = df["CELL_ID"].groupby(df["VILLES"])
ncpc = group_CELLID.count()
ncpc.sort_values(ascending= False)
```

**SPLITTING OF DATE_TIME COLUMN INTO DATE AND TIME COLUMNS**

```
all_days[["DATE", "TIME"]] = all_days.DATE_TIME.str.split(expand=True)
all_days
```

**SUM OF VALUES EACH DAY**

```
grouped_value = all_days["VALUE"].groupby(all_days["DATE"])
addition = grouped_value.sum()
addition.sort_values(ascending= False)
```

**NUMBER OF CELL_ID IN EACH REGION**

```
group_region = df["CELL_ID"].groupby(df["REGION"])
grouping = group_region.count()
grouping.sort_values(ascending= False)
```

**SUM OF VALUES OF EACH CELL_ID**

```
cell_group = all_days["VALUE"].groupby(all_days["CELL_ID"])
sort = cell_group.sum()
sort.sort_values(ascending=False)
```

```
all_days.rename(columns = {"SITE_ID": "SITE_CODE"}, inplace = True)
```

**MERGING OF DF AND ALL_DAYS TABLES**

```
merging = pd.merge(left= all_days, right= df, left_on= "SITE_CODE", right_on= "SITE_CODE")
merging
```

**CITIES WHERE MTN PRODUCTS WAS MOSTLY USED**

```
grouped_cities = merging["VALUE"].groupby(merging["VILLES"])
groupt = grouped_cities.sum()
groupt.sort_values(ascending=False)
```

**REGIONS WHERE MTN PRODUCTS WAS MOSTLY USED**

```
grouped_regions = merging["VALUE"].groupby(merging["REGION"])
groupregion = grouped_regions.sum()
groupregion.sort_values(ascending=False)
```

**USAGE OF MTN PRODUCTS IN EACH CITY FOR THE THREE DAYS**

```
merging.pivot_table(index=["DATE", "TIME", "VILLES"], aggfunc = 'count')
```
